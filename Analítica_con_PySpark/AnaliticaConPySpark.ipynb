##### 2. Cargue el archivo en un DataFrame de PySpark utilizando el método `spark.read.csv()` con la opción `header=True`.


```python
import pyspark
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder\
        .master("local[*]")\
        .appName('PySpark_training')\
        .getOrCreate()
```

```python
df = spark.read.csv("ventas.csv", header=True, inferSchema=True)
```

##### 3. Realice un análisis exploratorio de los datos mostrando el esquema (`printSchema()`) y las primeras filas (`show()`).


```python
df.printSchema()
```

```python
df.show()
```

##### 4. Limpie los datos eliminando registros nulos o inconsistentes utilizando `dropna()` o `fillna()`.


```python
df.dropna()  # Elimina registros nulos
```

##### 5. Ejecute operaciones analíticas como:
 - Agrupamiento (`groupBy`) por categoría o región.
 - Cálculo de promedios, máximos y mínimos (`agg`, `avg`, `max`, `min`).
 - Filtrado de registros mediante condiciones lógicas.
 

```python
df_grouped_by_Categoria = df.groupBy("Categoria").count()
df_grouped_by_Categoria.show()
```

```python
df_grouped_by_Region = df.groupBy("Region").sum("Unidades_Vendidas")
df_grouped_by_Region.show()
```

```python
# En Apache Spark, agg() se refiere a una función de agregación que se utiliza después de una operación groupBy() para calcular valores agregados


df.agg({'Precio': 'min'}).show()  # Precio Minimo
df.agg({'Precio': 'max'}).show()  # Precio Máximo
df.agg({'Precio': 'avg'}).show()  # Precio Promedio
```

##### 6. Guarde los resultados en un nuevo archivo CSV usando `write.csv()` dentro de una carpeta llamada 'output_csv'.

```python
df.toPandas().to_csv(
    r"output_csv/ventas_limpias.csv",
    index=False,
    encoding="utf-8-sig"
)
print("✅ Archivo exportado en CSV con Pandas")
```
